```python
import cv2
img_path = '/content/drive/My Drive/Augmented_images/IMG_4971_0_8461.jpeg'
img = cv2.imread(img_path)
for result in results:
    boxes = result.boxes.cpu().numpy()
    for i, box in enumerate(boxes):
        r = box.xyxy[0].astype(int)
        crop = img[r[1]:r[3], r[0]:r[2]]
        cv2.imwrite("/content/drive/My Drive/CustomObjDet/data/runs/detect/cropped2/" + str(i) + ".jpg", crop)
```
